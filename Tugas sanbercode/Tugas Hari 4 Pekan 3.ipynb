```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('ggplot')
df = pd.read_csv('vgsales.csv')
df.head()
```

<h2>Soal 1: Preparing Data</h2>

Manipulasi data tersebut dengan menggroupkan berdasarkan Genre, kemudian ambil rata2 penjualan untuk setiap Region Kecuali Global_Sales berdasarkan kategori Genre.

```python
#code here
df_group = df.groupby('Genre') [['NA_Sales','EU_Sales','JP_Sales','Other_Sales']].mean()
df_group.head()
```

<h2>Soal 2: Perbandingan Kuantitatif Barplot - Grouping Visualisasi Dengan Barplot</h2>

1. Visualisasikan gambar yang tadi kita manipulasi, untuk menghasilkan visualisasi seperti di bawah ini
2. Tuliskan apa insight yang bisa kamu dapat dari visualisasi tersebut

```python
#code here
df_transpose = df_group.transpose()
df_transpose.plot.bar(rot=0,figsize=(12, 8), width=0.8)


plt.title('Mean sales Video Game by Genre')
plt.xlabel('Region Sales')
plt.ylabel('Mean Sales')
plt.show()
```

Jawab:
Dari chart yang ditampilkan maka kita bisa menyimpulkan bahwa market penjualan video game dihampir semua genre North America memiliki market penjualan tertinggi

<h2>Soal 3: Perbandingan Kuantitatif Barplot - Stack Barplot</h2>

1. Visualisasikan gambar yang tadi kita manipulasi, untuk menhasilkan visualisasi seperti di bawah ini
2. Tuliskan apa insight yang bisa kamu dapat dari visualisasi tersebut

```python
#code here
fig, ax = plt.subplots(figsize=(12,8))
df_transpose = df_group.transpose()

ax.bar(df_transpose.index, df_transpose['Action'], label='Action')
ax.bar(df_transpose.index, df_transpose['Adventure'], bottom= df_transpose['Action'], label='Adventure')
ax.bar(df_transpose.index, df_transpose['Fighting'], bottom= df_transpose['Adventure']+ df_transpose['Action'], label='Figthing')
ax.bar(df_transpose.index, df_transpose['Misc'], bottom= df_transpose['Fighting']+ df_transpose['Adventure']+ df_transpose['Action'], label='Misc')
ax.bar(df_transpose.index, df_transpose['Platform'], bottom= df_transpose['Misc']+ df_transpose['Fighting']+ df_transpose['Adventure']+ df_transpose['Action'], label='Platform')
ax.bar(df_transpose.index, df_transpose['Puzzle'], bottom= df_transpose['Platform']+ df_transpose['Misc']+ df_transpose['Fighting']+ df_transpose['Adventure']+ df_transpose['Action'], label='Puzzle')
ax.bar(df_transpose.index, df_transpose['Racing'], bottom= df_transpose['Puzzle']+ df_transpose['Platform']+ df_transpose['Misc']+ df_transpose['Fighting']+ df_transpose['Adventure']+ df_transpose['Action'], label='Racing')
ax.bar(df_transpose.index, df_transpose['Role-Playing'], bottom= df_transpose['Racing']+ df_transpose['Puzzle']+ df_transpose['Platform']+ df_transpose['Misc']+ df_transpose['Fighting']+ df_transpose['Adventure']+ df_transpose['Action'], label='Role-Playing')
ax.bar(df_transpose.index, df_transpose['Shooter'], bottom= df_transpose['Role-Playing']+ df_transpose['Racing']+ df_transpose['Puzzle']+ df_transpose['Platform']+ df_transpose['Misc']+ df_transpose['Fighting']+ df_transpose['Adventure']+ df_transpose['Action'], label='Shooter')
ax.bar(df_transpose.index, df_transpose['Simulation'], bottom= df_transpose['Shooter']+ df_transpose['Role-Playing']+ df_transpose['Racing']+ df_transpose['Puzzle']+ df_transpose['Platform']+ df_transpose['Misc']+ df_transpose['Fighting']+ df_transpose['Adventure']+ df_transpose['Action'], label='Simulation')
ax.bar(df_transpose.index, df_transpose['Sports'], bottom= df_transpose['Simulation']+ df_transpose['Shooter']+ df_transpose['Role-Playing']+ df_transpose['Racing']+ df_transpose['Puzzle']+ df_transpose['Platform']+ df_transpose['Misc']+ df_transpose['Fighting']+ df_transpose['Adventure']+ df_transpose['Action'], label='Sports')
ax.bar(df_transpose.index, df_transpose['Strategy'], bottom= df_transpose['Sports']+ df_transpose['Simulation']+ df_transpose['Shooter']+ df_transpose['Role-Playing']+ df_transpose['Racing']+ df_transpose['Puzzle']+ df_transpose['Platform']+ df_transpose['Misc']+ df_transpose['Fighting']+ df_transpose['Adventure']+ df_transpose['Action'], label='Strategy')

plt.title('Mean sales Video Game by Genre')
plt.xlabel('Region Sales')
plt.ylabel('Mean Sales')

plt.legend()
plt.show()
```
